```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

STOPWORDS = set(stopwords.words('english'))
```

## Load data

```python
df = pd.read_csv('../data/LabeledText.csv', encoding='latin-1')
df = df[['Caption', 'LABEL']].copy()
df.head()
```

## Text cleaning

Steps:
1. Lowercase
2. Remove URLs
3. Remove non-alphabetic characters (keeps spaces)
4. Tokenize
5. Remove stopwords
6. Rejoin tokens

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\.\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [t for t in tokens if t not in STOPWORDS and len(t) > 1]
    return ' '.join(tokens)

df['clean'] = df['Caption'].apply(clean_text)
df[['Caption', 'clean', 'LABEL']].head(10)
```

```python
empty = (df['clean'].str.strip() == '').sum()
print(f'Empty after cleaning: {empty} / {len(df)}')
```

## Train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df['clean'], df['LABEL'],
    test_size=0.2,
    random_state=42,
    stratify=df['LABEL']
)

print(f'Train: {len(X_train)}  Test: {len(X_test)}')
print('\nTrain class distribution:')
print(y_train.value_counts(normalize=True).round(3))
print('\nTest class distribution:')
print(y_test.value_counts(normalize=True).round(3))
```
