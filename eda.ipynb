```python
import pandas as pd
import os
```

## Text class distribution

```python
df = pd.read_csv('data/LabeledText.csv', encoding='latin-1')
print(df.shape)
df.head()
```

```python
print(df.dtypes)
print(df.isnull().sum())
```

```python
label_col = df.columns[-1]
print(df[label_col].value_counts())
print(df[label_col].value_counts(normalize=True).round(3))
```

## Image class distribution

```python
img_root = 'data/Images'
counts = {cls: len(os.listdir(os.path.join(img_root, cls))) for cls in os.listdir(img_root)}
total = sum(counts.values())
for cls, n in counts.items():
    print(f'{cls}: {n} ({n/total:.1%})')
```
